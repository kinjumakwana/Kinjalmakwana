```python
# Write a python program to find the longest words.
def longest_word(filename):
    with open(filename, 'r') as infile:
              words = infile.read().split()
    max_len = len(max(words, key=len))
    return [word for word in words if len(word) == max_len]

print(longest_word('new_file.txt'))
```

```python
# Write a Python class named Rectangle constructed by a length and width and a method which will compute the area of a rectangle
class rectangle:
  def __init__(self,l,w):
    self.length= l
    self.width= w
  def rectangle_area(self):
    return self.length * self.width
  
new_rectangle=rectangle(20,25)
print(new_rectangle.rectangle_area())
```

```python
# Write a Python class named Circle constructed by a radius and two methods which will compute the area and the perimeter of a circle
class circle:
  def __init__(self,r):
    self.radius=r

  def circle_are(self):
    return 3.14*self.radius*self.radius
  
  def perimeter(self):
    return 2*self.radius*3.14

new_circle=circle(51)
print(new_circle.circle_are())
print(new_circle.perimeter())
```

> Explain Inheritance in Python with an example?

> Ans:

> ust like a parent-child relationship, inheritance works on derived classes relative to the base class. Every “Derived” class inherits from a “Base” class. The inheritance is represented in UML or Unified Modeling Language. It is a standard modeling language that includes an integrated set of diagrams to help developers specify, structure, and document software systems elements. 
Inheritance relationship defines the classes that inherit from other classes as derived, subclass, or sub-type classes. Base class remains to be the source from which a subclass inherits.

> Example are below

```python
# example inheritance 
# Single inheritance
class Employee:
  company="SSMP"
  def Showdetails(self):
    print("This is for Employee")

class Programmer(Employee):
  company="Google"
  language="Python"
  def datashow(self):
    print(f"Devloper go to {self.company} for experiance")

  def other(self):
    print(f"{self.language} was prefer to all students")

op=Programmer()
op.Showdetails()
print(op.company)
op.datashow()
op.other()
```

```python
# multilevel inheritance
class person:
  def TakeBreath(self):
    print("I am Breathing...")

class Employee(person):
  company="SSMP"
  def Showdetails(self):
    print("This is for Employee")

class Programmer(Employee):
  company="Google"
  language="Python"
  def datashow(self):
    print(f"Devloper go to {self.company} for experiance")

  def other(self):
    print(f"{self.language} was prefer to all students")
  
m=Programmer()
m.TakeBreath()
m.Showdetails()
print(m.language)
m.other()
```

```python
# multiple inheritance
class Employee:
  company="Google"
  ecode=120

class Freelancer:
  company="SSMP"
  level=0
  def incrememt(self):
    self.level=self.level+1

class Programmer(Employee,Freelancer):
  name="Meet"

op=Programmer()
print(op.level)
op.incrememt()
op.incrememt()
print(op.level)
print(op.company)
```
