- The cmp() function
- The set() function and == operator
- The sort() function and == operator

- The reduce() and map() function

The Python

```python
# I1. Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings.
def match_words(words):
  ctr = 0

  for w in words:
    if len(w) > 1 and w[0] == w[-1]:
      ctr += 1
  return ctr

print(match_words(['abc', 'xyz', 'aba', '1221']))
```

```python
# Write a Python program to remove duplicates from a list

def Remove(duplicate):
    final_list = []
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
     
# Driver Code
duplicate = [2, 4, 10, 20, 5, 2, 20, 4]
print(Remove(duplicate))
```

# What is tuple? Difference between list and tuple.

## List
List is a container to contain different types of objects and is used to iterate objects.

## Example

list = ['a', 'b', 'c', 'd', 'e']

Tuples
Tuple is also similar to list but contains immutable objects. Tuple processing is faster than List.

##Example

tuples = ('a', 'b', 'c', 'd', 'e')
Following are the important differences between List and Tuple.

Sr  Key	List	             Tuple
1	Type:	List is mutable.	Tuple is immutable.
2	Iteration:	List iteration is slower and is time consuming.	Tuple iteration is faster.
3	Appropriate for:	List is useful for insertion and deletion operations.	Tuple is useful for readonly operations like accessing elements.
4	Memory Consumption: List consumes more memory.	Tuples consumes less memory.
5	Methods: List provides many in-built methods.	Tuples have less in-built methods.
6	Error prone: List operations are more error prone.	Tuples operations are safe.

It is also a sequence data type capable of containing elements of multiple data types, and they are immutable. In simpler words, a tuple refers to a collection of various Python objects that stay separated by commas. A tuple is comparatively much faster than a list because it is static in nature.

```python
# Write a Python program to find the highest 3 values in a dictionary
from collections import Counter
dict1={
    "a":8945,
    "b":34,
    "c":20,
    "d":900,
}
k=Counter(dict1)
high=k.most_common(3)
for i in high:
  print(i[0],":",i[1])
```

```python
from heapq import nlargest
 
# Initial Dictionary
my_dict = {'A': 67, 'B': 23, 'C': 45,
           'D': 56, 'E': 12, 'F': 69}
 
print("Initial Dictionary:")
print(my_dict, "\n")
 
ThreeHighest = nlargest(3, my_dict, key = my_dict.get)
 
print("Dictionary with 3 highest values:")
print("Keys: Values")
 
for val in ThreeHighest:
    print(val, ":", my_dict.get(val))
```

```python
# Write a Python function to calculate the factorial of a number (a non-negative integer)
def calculater(x):
    fact=1
    for i in range(1,x+1):
      fact=fact*i
    return fact
y=calculater(int(input("Enter a number: ")))
print(y)
```

```python
# Write a Python function to check whether a number is perfect or not.
def perfect_numbers(N):
   sum = 0
   for i in range(1,N): 
      if(N%i == 0):
         sum = sum+i 
   return sum 

N = 6

if(N == perfect_numbers(N)): 
   print(N, "is a perfect number") 
else: 
   print(N, "is not a perfect number")
```

```python
# Write a Python function that checks whether a passed string is palindrome or not
def palindrome(n):
  if len(n) < 1:
    return True
  else:
    if n[0]==n[-1]:
      return palindrome(n[1:-1])
    else:
      return False
a=str(input("Enter a anything: "))
if palindrome(a)==True:
  print("This is palindrome")
else:
  print("This is not palindrome")
```

```python
def isPalindrome(string):
	left_pos = 0
	right_pos = len(string) - 1
	
	while right_pos >= left_pos:
		if not string[left_pos] == string[right_pos]:
			return False
		left_pos += 1
		right_pos -= 1
	return True
print(isPalindrome('aza')) 
```
