```python
# example

c = 1
info = dict()

while True:
  key = input('enter your key')
  info.setdefault(key)
  
  print('Do you want to add another key?: ')  
  status = input('0 to stop and 1 to continue: ')

  if status.isnumeric():
    status = int(status)
    if status:
      continue
    else:
      break

for k,v in info.items():
  print(k, v)
```

```python
# example

c = 1
info = dict()

keys = input('enter your keys (seperated by comma): ')

for k in keys.split(','):
  info.setdefault(k)

for k,v in info.items():
  print(k, v)
```
