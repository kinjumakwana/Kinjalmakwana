```python
# format(*args) - formatting string

welcome = "welcome to {}"
```

```python
welcome.format('python')
```

```python
welcome = "hello {}! welcome to {}"
```

```python
welcome.format('parth', 'python')
```

```python
"hello {}! welcome to {}".format('python', 'parth')
```

```python
# place format values using index number

"hello {1}! welcome to {0}".format('python', 'parth')
```

```python
# using keywords

"hello {name}! welcome to {lang}".format(lang='python', name='parth')
```

```python
# f-string/f-notation
name = 'parth'
lang = 'python'

f"hello {name}! welcome to {lang}"
```
