```python
# control statements - continue, break

for i in range(10):
  if i == 5:
    break
  print(i)
```

```python
for i in range(10):
  print(i)
  if i == 5:
    break
```

```python
for i in range(10):
  if i == 5:
    continue
  print(i)
```

```python
numbers = [5,6,4,7]
```

```python
for i in range(1, 11):
  print(i * 2)
```

```python
for n in numbers:
  print(n)
```

```python
for n in numbers:
  for i in range(1, 11):
    print(n * i)
```

```python
for n in numbers:
  for i in range(1, 11):
    print(f"{n} * {i} = {n * i}")
  print('--------------')
```

```python
inp = input('Enter your numbers to print table. (seperate by comma): ')
```

```python
inp
```

```python
s = inp.split(',')
```

```python
s
```

```python
for n in s:
  if n.isnumeric():
    for i in range(1, 11):
      print(f"{n} * {i} = {int(n) * i}")
    print('--------------')
```

```python
l = [5,4,56,4,6,4,4,1,1,2,3,2,1,4,1]
t = []
for i in l:
  if i not in t:
    t.append(i)
else:
  l = t.copy()
  del t
  print(l)
```
