# Conditional Statements

## 1. Simple if
> <pre>if condition:
    true_statement</pre>

## 2. if .. else
> <pre>if condition:
    true_statement
else:
    false_statement</pre>

## 3. if .. elif .. else ..
> <pre>if condition_1:
    true_statement_1
elif condition_2:
    true_statement_2</pre>

## 4. Nested if
> <pre>if outer_condition:
    true_statement
    if inner_condition:
        inner_true_statement
else:
    false_statement</pre>

## 5. Grouped if - using logical operators
> <pre>if condition_1 and condition_2:
    true_statement
else:
    false_statement</pre>

```python
# age verification

user_age = input('enter your age: ')
req_age = 18

if user_age.isnumeric():
  user_age = int(user_age)
  if user_age >= req_age:
    print('Congrations! You are eligible to apply for license.')
  else:
    print('Sorry! You are not eligible to apply for license.')
else:
  print('please enter age only.')
```

```python
user_age = input('enter your age: ')
req_age = 18

if user_age.isnumeric():
  user_age = int(user_age)
  if user_age >= req_age:
    print('Congrations! You are eligible to apply for license.')
  else:
    print(f'Sorry! You are not eligible to apply for license. Please try after {req_age - user_age} year(s).')
else:
  print('please enter age only.')
```

```python
user_input = input("enter random data: ")

# 'a' in 'apple'

for i in user_input:
  print(i, end=" ")
```

```python
nc = 0
sc = 0
for i in user_input:
  if i.isnumeric():
    # i = int(i)
    print(i, 'number', end=" ")
    nc += 1
  else:
    print(i, 'character', end=" ")
    sc += 1

print()
print('total num:', nc)
print('total char:', sc)
```
