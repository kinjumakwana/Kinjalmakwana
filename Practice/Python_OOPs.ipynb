# OBJECT-ORIENTED PROGRAMMING
## Object oriented programming is a technique or style which is associated with an object and a class.

## What is Class?
### Class is a collection of member variables and member methods. A Class just like a template, design or blueprint.

### Example:
- ### Class ->> Human Beings ->> Male and Female
- ### Varibales ->> Hands, Legs, Eyes, Ears and Heart.
- ### Functions ->> Walk, See, Talk, and Hear.

## What is Object?
### Object is an instance of a Class that holds a reference or address of member variables and member methods.

### Example:
- ### Kapil ->> OBJECT <<- Class Male

## Features of OOP
- ### Inheritance - It's allow to transmit data members from one class to another class.
### Example:

- ### Class HumanBeings ->> Inherit in both ->> Class Male and Class Female
- ### Encapsulation - It is called data hiding. It's a way to prevent accidental changes and unauthorised access to data members of a class. It's refer to binding the member variables and functions in single unit.
### Example:
- ### Class Male ->> Human voices can't be changed. Blood flow etc.

- ### Polymorphism - Means many forms. It allows us to define more than one way to do something.
- ### 1. Method Overloading ->> Compile time Polymorphism ->> Same name within same class and different number of arguments.
- ### 2. Method Overriding ->> Runtime Polymorphism ->> Same name, different class and same arguments, possible with Inheritance only.
### Example:
- ### Class HumanBeings ->> (hands) <<- walk ->> (legs) -- its called overloading
- ### Class HumanBeings ->> walk forward and walk backward -- changed the machenism of function, its called overriding.

#### --> __init__ - (It is reserved method in python classes known as  constructor. It is called when object created.)
#### --> self - (self parameter is a reference to the current object of a class. It's used to access variables that belongs to the class.)

## Inheritance - Way of transmission of member data from a class to another classes.
## types:
- ## Single Level - has a base class and a child class
- ## Multi-Level - has more than two classes
- ## Multiple - has two or more base classes and one child class
- ## Hierarchy - it is a tree based inheritance having one base class and muliple child classes
- ## Hybrid - it is mix of Multiple and Hierarchy inheritance.

```python
class Manager:
  managers = []
  def set_manager(self, name):
    self.name = name
    self.managers.append({'man_id': len(self.managers) + 1, 'man_name': self.name})

class Employee(Manager):
  data = []
  def set_employee(self, emp_name):
    self.emp_name = emp_name
    self.data.append({'emp_id': len(self.data) + 1, 'emp_name': self.emp_name})

  def assign_manager(self, man_id, emp_id):
    for i in self.data:
      if emp_id == i['emp_id']:
        i['man_id'] = man_id
    else:
      print('invalid emp id')
```

```python
class Manager:
  managers = []
  def set_manager(self, name):
    self.name = name
    self.managers.append({'man_id': len(self.managers) + 1, 'man_name': self.name})

class Employee(Manager):
  data = {}
  def set_employee(self, emp_name):
    self.emp_name = emp_name
    self.data.setdefault(
        len(self.data) + 1, {'emp_name': self.emp_name}
    )

  def assign_manager(self, man_id, emp_id):
    if emp_id in self.data:
        self.data[emp_id]['man_id'] = man_id
    else:
      print('invalid emp id')
```

```python
obj = Employee()
```

```python
obj.set_manager('kinjal')
obj.set_manager('nishi')
obj.set_manager('muskan')
```

```python
obj.set_employee('parth')
obj.set_employee('meet')
obj.set_employee('maulik')
obj.set_employee('kalpesh')
obj.set_employee('advait')
obj.set_employee('dhairya')
```

```python
obj.managers
```

```python
obj.data
```

```python
obj.assign_manager(1, 1)
```

```python
obj.data
```

```python
# multi-level

class Parents:
  def car(self):
    print('car from parents')

class Child(Parents):
  def bunglow(self):
    print('bunglow from child')

class GrandChild(Child):
  def bycycle(self):
    print('bycycle')
```

```python
obj = GrandChild()
```

```python
print(dir(obj))
```

```python
obj.car()
obj.bunglow()
obj.bycycle()
```

```python
# hierarchy

class Parents:
  def property(self):
    self.car = 'car from parents'
    self.bunglow = 'bunglow from parents'

class Son(Parents):
  def bycycle(self):
    print('bycycle from son')

class Daughter(Parents):
  def royal_enfield(self):
    print('bike from daughter')
```

```python
obj_son = Son()
obj_dau = Daughter()
```

```python
obj_son.property()

print(obj_son.car)
print(obj_son.bunglow)

obj_son.bycycle()
```

```python
obj_dau.property()

print(obj_dau.car)
print(obj_dau.bunglow)

obj_dau.royal_enfield()
```

```python
# hybrid

class Parents:
  def property(self):
    self.car = 'car from parents'
    self.bunglow = 'bunglow from parents'

class Child_1(Parents):
  def bycycle(self):
    print('bycycle from son')

class Child_2(Parents):
  def royal_enfield(self):
    print('bike from daughter')

class Son_of_Child_1(Child_1, Child_2):
  pass
```

```python
obj = Son_of_Child_1()
```

```python
print(dir(obj))
```

```python
# Polymorphism -- Method Overriding - using super() method

class Parents:
  def property(self):
    print('property from parents')
    car  = 'BMW'
    bunglow = ''
  
class Child(Parents):
  def property(self):
    print('property from child')
    bike = ''

class GrandChild(Child):
  def property(self):
    print('property from grand child')
    toys = ''
```

```python
obj = GrandChild()
```

```python
obj.property()
```

```python

class Parents:
  def property(self):
    print('property from parents')
    car  = 'BMW'
    bunglow = ''
  
class Child(Parents):
  def property(self):
    super().property()
    print('property from child')
    bike = ''

class GrandChild(Child):
  def property(self):
    super().property()
    print('property from grand child')
    toys = ''
```

```python
obj = GrandChild()
```

```python
obj.property()
```

```python
# using class name

class Parents:
  def property(self):
    print('property from parents')
    car  = 'BMW'
    bunglow = ''
  
class Child(Parents):
  def property(self):
    # super().property()
    Parents.property(self)
    print('property from child')
    bike = ''

class GrandChild(Child):
  def property(self):
    # super().property()
    Child.property(self)
    print('property from grand child')
    toys = ''
```

```python
obj = GrandChild()
```

```python
obj.property()
```
